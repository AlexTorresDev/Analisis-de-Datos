### Datos ánomalos

```python
# ENGINESIZE
def verificar_datos_anomalos(df, column_name, const):
    mean = df[column_name].mean()
    desv = df[column_name].std()

    lim_sup = mean + (desv * const)
    lim_inf = mean - (desv * const)

    print('Column name: ', column_name)
    print('Límite Superior: ', lim_sup)
    print('Límite Inferior: ', lim_inf)

    cant_sup = df[df[column_name] > lim_sup].shape[0]
    cant_inf = df[df[column_name] < lim_inf].shape[0]

    return cant_sup, cant_inf

print(verificar_datos_anomalos(df, 'ENGINESIZE', 3.6))
print(verificar_datos_anomalos(df, 'CYLINDERS', 3.6))
print(verificar_datos_anomalos(df, 'FUELCONSUMPTION_CITY', 3.6))
print(verificar_datos_anomalos(df, 'CO2EMISSIONS', 3.6))
print(verificar_datos_anomalos(df, 'FUELCONSUMPTION_COMB', 3.6))
```

```python
## Eliminar datos anomalos

df = df[df['FUELCONSUMPTION_CITY'] < 28.0]
```

```python
df
```

### Creación del modelo

#### Selección de features

```python
features = ['ENGINESIZE', 'CYLINDERS', 'FUELCONSUMPTION_CITY', 'FUELCONSUMPTION_COMB', 'FUELCONSUMPTION_HWY']
label = ['CO2EMISSIONS']

X = df[features]
y = df[label]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15)

print(X_train.shape)
print(X_test.shape)
```

```python
# Entrenamiento
modelo = LinearRegression().fit(X_train, y_train)

# Aplicación
modelo.score(X_test, y_test)
```

```python
# Uso del modelo

print(df[df['ENGINESIZE'] > 7.5])

ejemplo = pd.DataFrame({'ENGINESIZE': [8.4], 'CYLINDERS': [10], 'FUELCONSUMPTION_CITY': [20], 'FUELCONSUMPTION_COMB': [16.9], 'FUELCONSUMPTION_HWY': [13.0]})
modelo.predict(ejemplo)
```
